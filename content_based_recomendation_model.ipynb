```python
import pandas as pd
import numpy as np
import random

locations = pd.read_csv("city.csv")
products_df = pd.read_csv("flipkart_prod.csv")

num_rows = len(products_df)

print(num_rows)

city_names = locations['City'].tolist()
state_names = locations['State'].tolist()
latitude_values = locations['Lat'].tolist()
longitude_values = locations['Long'].tolist()

# Generate random lists of city and state names
random_city_names = random.choices(city_names, k=num_rows)
random_state_names = [state_names[city_names.index(city)] for city in random_city_names]
random_latitude_values = [latitude_values[city_names.index(city)] for city in random_city_names]
random_longitude_values = [longitude_values[city_names.index(city)] for city in random_city_names]

products_df['City'] = random_city_names
products_df['State'] = random_state_names
products_df['Latitude'] = random_latitude_values
products_df['Longitude'] = random_longitude_values

products_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3,  max_features=None,
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')

# Filling NaNs with empty string
products_cleaned_df['product_specifications'] = products_cleaned_df['product_specifications'].fillna('')
```

```python
tfv_matrix = tfv.fit_transform(products_cleaned_df['product_specifications'])
```

```python
def give_rec_spec(product_name, sig=sig):
    # Get the index corresponding to unique Id
    uniqueId = products_cleaned_df.loc[products_cleaned_df['product_name']==product_name,'uniq_id'].values[0]
    idx = indices[uniqueId]
    # Get the pairwsie similarity scores
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Product indices
    product_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return products_cleaned_df.iloc[product_indices]
```

```python
# Testing our content-based recommendation system with the seminal film Spy Kids
arr=give_rec_spec('Audeze Lcd2 In Bamboo, High Quality Planar Magnetic Headphone Headphones')
```

```python
import math
from sklearn.metrics.pairwise import haversine_distances
from math import radians

# Function to calculate the distance between two points using Haversine formula
def haversine_sklearn(lat1, lon1, lat2, lon2):
    coords_1 = [radians(lat1), radians(lon1)]
    coords_2 = [radians(lat2), radians(lon2)]
    result = haversine_distances([coords_1, coords_2])
    distance = result[1][0] * 6371000  # Multiply by Earth's radius to get distance in meters
    return distance

def give_rec_loc(user_latitude, user_longitude):
    distance = products_cleaned_df.apply(lambda row: haversine_sklearn(user_latitude, user_longitude, row['Latitude'], row['Longitude']), axis=1)
    products_cleaned_df['score'] = 1 / (distance + 1)  # Assign a score based on distance

    top_products = products_cleaned_df.sort_values(by='score', ascending=False).head(10)

    return top_products
```
